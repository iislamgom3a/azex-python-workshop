# Strings

- A string is a collection of characters closed within single or double quotation marks (immutable).
- You can update an existing string by (re)assigning a variable to another string.
- python does not support a character type; these are treated as strings of length one.
- Python accepts single ('), double (") and triple (''' or """) quotes to denote string literals.
- String indexes starting at 0 in the beginning of the string and working their way from -1 at the end

````python
str1 = "I am a string"
str2 = "I am a string also"
str3 = """
I am a multi-line string 
"""
# ------------ Indexing ---------- #
first = "Islam"
print(first[0])
print(first[1])
print(first[-1])
````

## Slicing 

> str[start_end : end_index] NOTES : The character at the end index in the string, will not be included in the substring obtained through this method.



````python
str = "Islam Gomaa"
print(str[0:4])
print(str[3 : len(str)])  # slicing to the end
print(str[3:])  # another way to slice to the end of the string
# --------- slicing with a step ----------#
print(str[2::2])  # step = 2
print(str[5:2:-1])  # reverse steps
````

## Else Statement

- The else keyword catches anything which isn't caught by the preceding conditions.
- The else statement is executed when the if condition (and any elif conditions) evaluate to False.

````python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
````

````python
# else without elif
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
````

# Loops

> Python has two primitive loop commands: `while` loops , `for` loops


## while loops

````python
i = 1
while i < 6:
  print(i)
  i += 1
````

````python
# break Statement
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

````python
# continue Statement
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
````

## For Loops

````python
for x in "banana":
  print(x)
````

##  range() Function

````python
for x in range(6):
  print(x)
````

````python
for x in range(2, 6):
  print(x)
````
